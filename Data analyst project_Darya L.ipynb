### 4. Функция исследования DAU

```python
def DAU_analysis (file_name):
    DAu = pd.read_excel(file_name)
    DAu['date'] = pd.to_datetime(DAu['date'])
    DAU = DAu.set_index('date')
    
    DAU.plot(figsize = (15, 5))
    plt.title('Изменение DAU'), plt.xlabel('Дата'), plt.ylabel('Уникальные пользователи')
    plt.legend(bbox_to_anchor=(1.135, 0.7), frameon = True, facecolor = 'white')
    
    DAU_ = DAU.copy()
    DAU_['Total'] = DAU_.loc[:, DAU_.columns != 'Total'].sum(axis = 1)
    
    for column in DAU_:
        DAU_[column] = DAU_[column]/DAU_['Total']
    DAU_.loc[:, DAU_.columns != 'Total'].plot.area(figsize = (15, 5))
    plt.title('График с накоплением истории изменения процента DAU'), plt.xlabel('Дата'), plt.ylabel('Доля от общего DAU, %')
    plt.style.use('seaborn-pastel')
    plt.legend(bbox_to_anchor=(1.135, 0.7), frameon = True)
            
    return DAU, plt.show(), DAU_, plt.show()
```

```python
DAU_analysis('hw_lesson03_dau_channel.xlsx')
```

```python
DAU_analysis('hw_lesson03_dau_region.xlsx')
```

### 4. Построение воронки

```python
events.head()
```

```python
sales_funnel = events.groupby('event_name').agg({'event_name': 'count', 'user_id': 'nunique'})
sales_funnel.columns = ['Число событий', 'Уникальные пользователи']
sales_funnel = sales_funnel.sort_values(by = 'Уникальные пользователи', ascending = False)
sales_funnel
```

```python
sales_funnel['% от первого шага воронки'] = sales_funnel['Уникальные пользователи'] / sales_funnel['Уникальные пользователи']['login']
sales_funnel
```

```python
sales_funnel.style.format({'% от первого шага воронки': '{:.2%}'})
```

```python
colors = ['indianred' if x < 0.65 else 'pink' for x in sales_funnel['% от первого шага воронки'].values]
import plotly
from plotly import graph_objects as go
plotly.offline.init_notebook_mode(connected=True)

fig = go.Figure()

fig.add_trace(go.Funnel(
                        y = sales_funnel.index.values,
                        x = sales_funnel['Уникальные пользователи'],
                        textinfo = "value+percent initial",opacity = 0.65, marker = {"color": colors }))

fig.update_layout(
                    title = 'Воронка по событиям в Clash of Clowns',
                    yaxis_title = 'Шаги воронки',
                  )
fig.update_layout(margin=dict(l=0, r=0, t=25, b=0))

fig.update_layout(plot_bgcolor='white')

fig.show()
```

**Выводы и наблюдения:**
- Резкая потеря пользователей начинается после события - прохождение обучения: в игру продолжают играть 80% людей от людей, прошедших обучение и это составляет 57% от количества людей, запустивших игру. Значит специфика/ механика игры нравится не всем людям, решившим попробовать ее посмотреть.
- Повышают уровень персонажа только 38% людей от поигравших в Clash of clowns впервые.
- Покупку совершили только 7.8% людей, возвысивших персонажа до нового уровня, при этом это всего 2% от людей запустивших игру.
- Приглашают друга в игру только 1% людей от количества запустивших игру. Это говорит о том, что только 1% людей  от общего количества людей, которых игра привлекла, и они решили просто ее запустить, действительно игра понравилась, поэтому они захотели поиграть с другом/ посоветовать другу игру. #### Рассчитаем необходимую длительность A/B-теста.

```python
def get_sample_size(mean, # среднее метрики
                    std, # среднеквадратическое отклонение (СКО) метрики         
                    relative_practical_significance, # минимальный эффект - процент изменений среднего метрики, который мы хотим обнаружить
                    alpha, # уровень значимости (вероятность обнаружить эффект случайно)
                    power # мощность (вероятность НЕ отбросить случайно реально существующий эффект)
                   ):
    
    z = stats.norm.ppf(1 - alpha / 2) + abs(stats.norm.ppf(1 - power))
    d = mean * relative_practical_significance
    
    return math.ceil(2 * ((std * z / d) ** 2))
```

```python
#Для нахождения среднего и СКО воспользуемся формулами биноминального распределения
p = history_funnel_events['Unique users']['tutorial_completed']/history_funnel_events['Unique users']['login']
mean_of_successful_tutorial = p
std_of_successful_tutorial = np.sqrt(p * (1 - p))
print('Средняя вероятность прохождения обучения успешно: {:.2%}, СКО: {:.2f}'.format(mean_of_successful_tutorial, std_of_successful_tutorial))
```

```python
alpha = 0.05
power = 0.8
relative_practical_significance = 0.038
```

```python
sample_size = get_sample_size(mean_of_successful_tutorial, std_of_successful_tutorial, relative_practical_significance, alpha, power)
sample_size
```

```python
def retention (profiles, sessions, grouping_attribute, age_limit):
    retention_raw = profiles.merge(sessions[['user_id','dt']], on = 'user_id', how = 'left')
    retention_raw['age'] = (retention_raw['dt']-retention_raw['first_date']).dt.days
    retention = retention_raw.pivot_table(index = grouping_attribute, columns = 'age', values = 'user_id', aggfunc = 'nunique').fillna(0)
    retention = retention.div(retention[0], axis = 0)
    report = retention.drop(columns = [0])
    ax = report.T.plot(grid = True)
    display(report.style.format('{:.2%}'))
    import matplotlib.ticker as mtick
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))
    plt.xlim([0, age_limit])
    plt.style.use('ggplot')
    plt.title('Удержание'), plt.xlabel('Дни с момента регистрации'), plt.ylabel('% пользователей');
    plt.show()
    
    heatmap = report.copy()
    import seaborn as sns
    y = sns.heatmap(heatmap, cmap= 'Spectral')
    plt.xlim([0, age_limit])
    
    return 
```

```python
def LTV (profiles, purchses, grouping_attribute, age_limit):
    ltv_info = profiles.merge(purchases[['user_id', 'dt', 'revenue']], on = 'user_id', how = 'left')
    ltv_info['age'] = (ltv_info['dt'] - ltv_info['first_date']).dt.days
    ltv_info.query('buyer == True')
    ltv = ltv_info.pivot_table(index = grouping_attribute, columns = 'age', values = 'revenue', aggfunc = 'sum').fillna(0)
    ltv = ltv.cumsum(axis = 1)
    cohort_sizes = profiles.groupby(grouping_attribute).agg({'user_id': 'nunique'})
    cohort_sizes.columns = ['cohort_size']
    report = cohort_sizes.merge(ltv, on = grouping_attribute)
    report = report.div(report['cohort_size'], axis = 0)
    report_ltv = report.drop(columns = 'cohort_size')
    display(report_ltv.style.format('{:.4f}'))
    report_ltv.T.plot(grid = True)
    plt.style.use('ggplot')
    plt.xlim([0, age_limit])
    plt.title('LTV'), plt.xlabel('Дни с момента регистрации'), plt.ylabel('LTV, $')
    plt.show()
    
    heatmap = report_ltv.copy()
    import seaborn as sns
    y = sns.heatmap(heatmap, cmap= 'Spectral')
    plt.xlim([0, age_limit])
    return 
```
